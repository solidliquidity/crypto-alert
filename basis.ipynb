```python
import ccxt
import pandas as pd

# Initialize Binance API
binance = ccxt.binance()
binance_futures = ccxt.binance({'options': {'defaultType': 'future'}})

# Fetch available perpetual futures symbols on Binance
futures_markets = binance_futures.load_markets()
perpetual_symbols = [symbol for symbol in futures_markets if "/USDT" in symbol and futures_markets[symbol]['contract']]

# Create an empty DataFrame
df = pd.DataFrame(columns=["Symbol", "Spot Price", "Perpetual Price", "Basis", "Funding Rate (%)"])

i = 0
while i < 100:
    ticker = perpetual_symbols[i].split(":")[0]  # Ensure correct ticker format

    try:
        # Fetch latest spot price
        spot_price = binance.fetch_ticker(ticker)['last']

        # Fetch latest perpetual futures price
        perp_price = binance_futures.fetch_ticker(ticker)['last']

        # Fetch latest funding rate
        funding_data = binance_futures.fapiPublicGetFundingRate({"symbol": ticker.replace("/", ""), "limit": 1})
        latest_funding_rate = float(funding_data[0]['fundingRate']) * 100  # Convert to percentage

        # Calculate basis (Perp - Spot)
        basis = perp_price - spot_price

        # Append to DataFrame
        df.loc[len(df)] = [ticker, spot_price, perp_price, basis, latest_funding_rate]

    except Exception as e:
        print(f"Error fetching data for {ticker}: {e}")

    i += 1

# Display the DataFrame
print(df)
```
